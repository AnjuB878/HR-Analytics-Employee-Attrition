## 📊 STEP 3: Exploratory Data Analysis (EDA)

### 1. Overall Attrition Rate

```python
attrition_rate = df['Attrition'].mean() * 100
print("Attrition Rate:", attrition_rate)
```

### 2. Attrition Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Attrition', data=df)
plt.title("Attrition Distribution")
plt.show()
```

### 3. Attrition by Department

```python
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()
```

### 4. Attrition vs Salary

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Attrition vs Salary")
plt.show()
```

### 5. Attrition vs Overtime

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("Attrition vs Overtime")
plt.show()
```

### 6. Attrition vs Job Satisfaction

```python
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title("Attrition vs Job Satisfaction")
plt.show()
```

### 7. Attrition vs Years at Company

```python
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', bins=10)
plt.title("Attrition vs Years at Company")
plt.show()
```
